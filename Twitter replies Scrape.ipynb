## Load the dataset which contains the Tweet ID's

```python
data  = pd.read_csv("liMiquela_tweets_phase2.csv")
data.drop('Unnamed: 0', axis = 1 , inplace= True)
total_tweets = len(data)
```

```python
data = data.fillna(0)
data.tail(5)
```

```python
def close_browser(browser):
    browser.close()
    return
```

### Capture followers responses against each tweet based on Tweet ID

```python
tweet          = []
original_tweet = [] 
reply_tweets   = []

#total_tweets = 3
i=0
    
for i in range(total_tweets):
    browser      = webdriver.Chrome(ChromeDriverManager().install())
    browser.wait = WebDriverWait(browser, 5)
    browser.get("https://twitter.com/lilmiquela/status/" + str(data.Tweet_ID[i]))
    browser.maximize_window()
    browser.wait = WebDriverWait(browser, 5)
    body = browser.find_element_by_tag_name('body')

    last_height = browser.execute_script("return document.body.scrollHeight")
    
    while True:
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = browser.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    #tweets = browser.find_elements_by_class_name('ThreadedConversation-tweet')
    tweets = browser.find_elements_by_class_name('js-tweet-text-container')
    num_of_tweets = len(tweets)
    
    j=0
    for j in range(num_of_tweets):
        
        reply_tweets.append(tweets[j].text.replace('\n',''))
        tweet_ID.append(str(data.Tweet_ID[i]))
        original_tweet.append(str(data.Text[i]))    
                
    close_browser(browser)
```

```python
df_reply = pd.DataFrame({'Tweet_ID' : tweet_ID, 'Original_Tweet_Text' : original_tweet, 'Reply_Tweets' : reply_tweets})
df_reply.fillna(0)
df_reply.to_csv('reply_phase2.csv')
df_reply.head(6)
```
